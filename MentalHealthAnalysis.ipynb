```python
'''
count:-

Definition: The count statistic represents the number of non-null entries in the column.
Usage: It helps to understand how many valid data points you have in the column.
Example: If a column has 7 entries and 2 of them are missing (NaN), the count will be 5.


unique:-

Definition: The unique statistic represents the number of distinct values in the column.
Usage: It provides information on the diversity of values in the column.
Example: If a column contains the values ['A', 'B', 'A', 'C', 'B'], the unique will be 3, as there are three distinct values ('A', 'B', and 'C').

top:-
Definition: The top statistic represents the most frequent value (mode) in the column.
Usage: It shows which value appears the most frequently in the column.
Example: If a column contains the values ['A', 'B', 'A', 'C', 'A'], the top will be 'A', as 'A' appears most frequently.

freq:-
Definition: The freq statistic represents the frequency (number of occurrences) of the most frequent value (mode) in the column.
Usage: It tells you how many times the most frequent value appears in the column.
Example: If a column contains the values ['A', 'B', 'A', 'C', 'A'], the freq will be 3, as 'A' appears 3 times.





'''
```

```python
def Counsplot(category):
    
    custom_colors = [
    '#e5c3c6',  
    '#cad2a7',  
    '#f96161', 
    '#bcd2d0',  
    '#d0b783',  
    '#a3d2ca',  
    '#f2b8ff',     
  ][:len(Mental_df[category].unique())]

    plt.figure(figsize=(10,8))
    sns.countplot(data=Mental_df,x= category,hue=category,palette=custom_colors)
    plt.xlabel(category)
    plt.xticks(rotation=45)
    plt.ylabel("Count")
    plt.title("Count of each categories")
    plt.show()

     # calculate counts and percentages
    category_counts = Mental_df[category].value_counts()
    # print(f'categoriesCounts---->{category_counts}')

    category_percentages = Mental_df[category].value_counts(normalize=True)
    # print(f'category_percentages----->{category_percentages}')

    # combine counts and percentages into a DataFrame
    res_df = pd.DataFrame({
        'Count': category_counts,
        'Proportion': category_percentages
    })
    res_df['Count'] = res_df['Count'].apply("{:,}".format)


    res_df['Proportion'] = res_df['Proportion'].apply("{:.1%}".format)
    return res_df
```

```python
# .apply("{:,}".format): The apply method is used to apply a function to each element of the selected column. In this case, the function being applied is "{:,}".format.

# "{:,}".format: This is a string format method that inserts commas as thousands separators into a number. For example:
# 1000 becomes 1,000
# 1000000 becomes 1,000,000




# .apply("{:.1%}".format): The apply method is used to apply a function to each element of the selected column. In this case, the function being applied is "{:.1%}".format.

# "{:.1%}".format: This is a string format method that converts a number to a percentage with one decimal place. The number is multiplied by 100 and a percent sign is added.
# For example:
# 0.1234 becomes 12.3%
# 0.5 becomes 50.0%
```

```python
for i in Mental_df.columns:
    if Mental_df[i].nunique()>2:
        print(Counsplot(i))
```

```python
def top_10_bar(category,x_rotation=45):

    custom_colors = [
    '#e5c3c6',  
    '#cad2a7',  
    '#f96161', 
    '#bcd2d0',  
    '#d0b783',  
    '#a3d2ca',  
    '#f2b8ff',     
  ][:len(Mental_df[category].unique())]

    #get top 10 of the category 
    top_10_bar=Mental_df[category].value_counts().head(10).reset_index()

    #plot the bar plot
    sns.barplot(data=top_10_bar,x=category,y="count",palette=custom_colors,hue=category)
    plt.title(category)
    plt.xticks(rotation = x_rotation)
    plt.xlabel(category)
    plt.ylabel("Count")
    plt.show()
    

    
    
```

```python
'''
reset_index(): Resets the index of the DataFrame to the default integer index and optionally converts the current index into columns.
drop=True: Drops the old index instead of converting it into a column.
inplace=True: Modifies the DataFrame in place.
'''
```

```python
for column in Mental_df.columns:
    if Mental_df[column].nunique()>2:
        top_10_bar(column)
```

```python
'''
groupby("gender"): Groups the DataFrame by the "gender" column.
["occupation"]: Selects the "occupation" column from the grouped DataFrame.
Result: A GroupBy object that allows you to perform group-wise operations on the "occupation" column within each gender group.
'''
```

```python
gender_data=(
    Mental_df.groupby("gender")["occupation"].value_counts().reset_index(name="count").sort_values(by="count",ascending=False)
)

custom_colors = [
    '#e5c3c6',  
    '#cad2a7',  
    '#f96161', 
    '#bcd2d0',  
    '#d0b783',  
    '#a3d2ca',  
    '#f2b8ff',     
  ][:len(Mental_df["gender"].unique())]
#plot the graph

sns.barplot(x="occupation",y="count",hue="gender",palette=custom_colors,data=gender_data)
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.title("Occupational Count of each Gender")
```

```python
'''

Mental_df by the values in the "gender" column and then select the "occupation" column from these groups. This operation is the first step in performing group-wise operations on the "occupation" column based on the unique values in the "gender" column.

Here's a breakdown of what each part of this line does:

Mental_df.groupby("gender"):

This part of the code groups the DataFrame Mental_df by the values in the "gender" column. It creates a GroupBy object, which is a special type of DataFrame that contains information about the groups.
Each unique value in the "gender" column will define a group. For instance, if the "gender" column has values 'Male' and 'Female', the DataFrame will be divided into two groups: one for 'Male' and one for 'Female'.
["occupation"]:

After grouping by "gender", this part selects the "occupation" column from the grouped DataFrame. This means you're focusing on the "occupation" column within each gender group.




'''
```
