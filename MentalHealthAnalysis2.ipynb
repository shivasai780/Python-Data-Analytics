```python
Education_data = Mental_data.groupby(["v5","rdep"],observed=True).size().reset_index(name="Count")
```

```python
Education_data["v5"].unique()
```

```python
#Here the Category is heavy ,so we can create another column
education_bins = {
    '1-10': ['10', '10 PASS', '10th std', 'CLASS 10', '9', '9 PASS', '9th std', '8', '8 PASS', '8TH', '8th std', '7', '7TH', '7th std', '6', '6TH', '6th std', '5', '5TH', '5th std', '4', '4TH', '4th std', '3', '3rd std', '2nd puc', 'Primary', 'Upper Primary'],
    '10-12': ['12', '2nd puc', 'CLASS 12', 'HSC', 'HS', 'Higher Secondary'],
    'BTech': ['B TECH', 'B Tech', 'B.TECH', 'B.tech', 'BE', 'B.E', 'B. Engineering'],
    'PostGraduationorMTech': ['M TECH', 'MBA', 'MCA', 'M.COM', 'M. Com', 'MSc', 'MSc', 'MS', 'MFC', 'PHD', 'POST DOCTORATE', 'POST GRADUATE']
}#

# Create a function to map the qualifications to bins
def categorize_education(qualification):
    for category, values in education_bins.items():
        if qualification in values:
            return category
    return 'Other'  # In case there are qualifications not covered in the bins

Education_data["Education Category"]=Education_data['v5'].apply(categorize_education)
```

```python
Education_data=Education_data.groupby(["Education Category","rdep"],observed=True).size().reset_index(name="count")
```

```python
Education_data
```

```python
plt.figure(figsize=(12,8))
figGen=sns.barplot(x="Education Category",y="count",data=Education_data,hue="rdep")
figGen.set_xlabel="Education"
figGen.set_ylabel="count"
figGen.set_title("Education  analysis of Mental illness")
```
