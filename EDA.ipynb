### 3 Sigma Technique

```python
data=pd.read_csv("./EDA-main/Data/raw_sales.csv")
```

```python
data.head(10)
```

```python
def detectAnamolies(data):

    #anamolies 
    anamolies=[]

    #Set an upper limit and lower limit to 3 Standard Deviation
    random_data_std=statistics.stdev(data)
    random_data_mean=statistics.mean(data)
                                    
    #3 standard deviation
    anamoly_cuttoff=random_data_std*3

    lower_limit=random_data_mean - anamoly_cuttoff
    upper_limit=random_data_mean + anamoly_cuttoff

    #Generate Outliers

    for outlier in data:
        if outlier>upper_limit or outlier<lower_limit:
            anamolies.append(anamolies)



    return anamolies
```

```python
df_Churn.head(10)
```

```python
df_Churn["Age"].min()
```

```python
df_Churn["Age"].max()
```

```python
labels=["0-20","21-40","41-60","Above 61"]
bins=[0,20,40,60,100]#it is between 0-20,21-40,41,
df_Churn["Age_bins"]=pd.cut(df_Churn.Age,bins,labels=labels,include_lowest=True)

# The labels list contains the names of the categories for the age intervals.
# The bins list defines the intervals: [0, 21), [21, 41), [41, 61), and [61, 100].

# Ages between 0 and 20 (inclusive of 0, exclusive of 21) are labeled as "0-20".
# Ages between 21 and 40 (inclusive of 21, exclusive of 41) are labeled as "21-40".
# Ages between 41 and 60 (inclusive of 41, exclusive of 61) are labeled as "41-60".
# Ages between 61 and 100 (inclusive of 61, exclusive of 100) are labeled as "Above 61".

# 
# The value 100 is the upper boundary for the last interval, ensuring that any Age value up to and including 100 falls within the last bin.


# Purpose of bins
# The bins list defines the intervals for categorizing the Age values. Each interval (or bin) is defined by two numbers, a lower boundary and an upper boundary. The pd.cut function will use these boundaries to determine which bin each Age value falls into.
```

```python
df_Churn.head(30)
```

```python
df_Churn.Age_bins.value_counts()
```

```python
def add_labels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
```

```python
#making the bar chart on the data

plt.bar(labels,df_Churn.Age_bins.value_counts())
add_labels(labels,df_Churn.Age_bins.value_counts())#here bins are integers and 
plt.title("AGE COUNT ")
plt.xlabel("Age Bins")
plt.ylabel("Age Count")
```
